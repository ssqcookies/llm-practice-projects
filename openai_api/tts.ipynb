# OpenAI TTS（文字转语音）模型介绍
    OpenAI 的 Text-to-Speech（TTS）是官方提供的文字配音服务，核心功能是将文本转换为自然流畅的语音，是 OpenAI 多模态生态的重要组成部分。
## 核心能力与应用场景
    朗读书面博客、文章内容
    多语言口语音频制作
    流媒体实时语音输出（如语音助手、直播播报）
## 核心配置项详解
    1. 语音音色（6 种）
    OpenAI TTS 内置 6 种不同风格的语音，覆盖多种场景：
    音色名	风格特点	适用场景
    alloy	中性、平衡的音色	通用场景、新闻播报
    echo	清晰、偏男性的音色	教程讲解、有声书旁白
    fable	温暖、有叙事感的音色	故事朗读、小说配音
    onyx	低沉、稳重的音色	企业播报、正式场合
    nova	明亮、偏女性的音色	语音助手、儿童内容
    shimmer	柔和、温暖的音色	有声书、情感类内容
    2. 输出格式
    默认输出格式为 mp3，同时支持多种格式适配不同场景：
    格式	特点与适用场景	
    mp3	通用压缩格式，兼容性最强	绝大多数场景，如网页播放、音频分享
    opus	低延迟、流媒体优化	实时语音通话、直播、视频会议
    aac	高效压缩，主流平台适配	YouTube、iOS/Android 应用、短视频配音
    flac	无损压缩，音质保留完整	音频存档、专业制作、音乐后期
    wav/pcm	无压缩原始音频，低延迟	实时语音处理、本地开发、音频编辑
    3. 语言支持
    OpenAI TTS 对英语发音做了深度优化，同时支持多语言文本转语音，但非英语（尤其是中文）的发音自然度、咬字清晰度不如英文。

# 智谱对应模型：智谱语音合成（TTS）服务
    智谱提供了原生的语音合成 API，是对标 OpenAI TTS 的国内替代方案，对中文场景做了深度优化，国内可直接调用，无需代理和海外付费。
   ## 核心能力与特点
    中文优化：针对普通话、方言（如粤语）、中文文本的自然度、语调、咬字做了专门优化，比 OpenAI TTS 更适合中文场景。
    多音色选择：提供多种风格的音色（如温柔女声、沉稳男声、童声、旁白音），覆盖教育、有声书、客服、短视频配音等场景。
    低延迟：国内直连，响应速度快，支持实时语音合成，适配语音助手、直播等低延迟场景。
    多格式支持：支持 MP3、WAV 等主流音频格式，兼容绝大多数播放场景。
##  核心参数与对应关系
    OpenAI TTS 参数	智谱语音合成对应说明
    model	智谱提供标准 / 高清版本，如 tts-1 对应标准模式
    voice	多种中文音色可选，覆盖不同性别、年龄、风格
    input	支持中文、英文等多语言文本，中文适配度更高
    response_format	支持 MP3、WAV 等格式，默认 MP3
    speed   生成音频的速度，默认速度为 1.0。

```python
from zhipuai import ZhipuAI
import os
from dotenv import load_dotenv
load_dotenv()

# 初始化智谱客户端
client = ZhipuAI(api_key=os.getenv("ZHIPUAI_API_KEY"))

response = client.audio.speech(
    model="glm-tts",  # 智谱语音合成模型
     voice="female",
    input="好饿啊好饿啊，饿死我了",
    response_format="wav",
    encode_format='base64',
)

# 保存音频文件
speech_file_path = "./audio/zhipu_tts_output.mp3"

response.stream_to_file(speech_file_path)
```
