```python
import os
from dotenv import load_dotenv
from langchain_openai import ChatOpenAI

# 自动加载 .env 文件里的所有 key
load_dotenv()

llm = ChatOpenAI(
    model="glm-4.5-air",
    api_key=os.getenv("ZHIPUAI_API_KEY"),
    base_url=os.getenv("ZHIPUAI_BASE_URL")
)


# 提问
response = llm.invoke("你好，请介绍一下你自己")

# 输出结果
print("智谱回答：", response)
```
